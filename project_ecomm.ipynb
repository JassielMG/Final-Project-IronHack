# El peso del pedido retrasa la entrega?
* De acuerdo a la prueba de hipotesis no hay una relaci√≥n entre los dos eventos, por lo que podemos decir que los dos eventos son independientes
* EL peso del producto no retrasa la entrega, dependera de otros eventos si la entrega se retrasa.
---



```python
# tablas que relacione:
# orders_df.head(3)
# products_df.head(3)
# items_df.head(3)

orders_items = pd.merge(orders_df,items_df,on='order_id',how='left')
orders_items_products = pd.merge(orders_items,products_df,on='product_id',how='left')

time_estimated = pd.to_datetime(orders_items_products['order_estimated_delivery_date']) - pd.to_datetime(orders_items_products['order_purchase_timestamp'])

time_estimated = time_estimated.dt.days.fillna(time_estimated.dt.days.median())

product_weight_g = orders_items_products['product_weight_g'].fillna(orders_items_products['product_weight_g'].median())
```

```python
# Realiz√© una prueba de hipotesis de dos muestras para encontrar la relaci√≥n entre la columna del peso de producto y el tiempo estimado de entrega

from scipy.stats import ttest_rel

ttest_rel(time_estimated,product_weight_g)
```

```python
# De acuerdo al estadistico y al p value no hay una relaci√≥n significativa en las dos muestras
```

```python
import itertools
pdq = list(itertools.product(p, d, q))
pdq
```

```python
season_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]

print(f'Sarimax parameters: {pdq[1]} x {season_pdq[1]}')
```

```python
for param in pdq:
  for param_estacional in season_pdq:
    try:
      mod = sm.tsa.statespace.SARIMAX(
          orders_time['sales'].resample('M').sum(),
          order=param,
          seasonal_order=param_estacional,
          enforce_stationarity=False, 
      )
      res = mod.fit()
      
      print(f'SARIMAX {param} x {param_estacional}, 12 - AIC: {res.aic}')

    except:
      continue
```
